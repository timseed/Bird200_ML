```python
import pandas as pd
from PIL import Image 
from glob import glob

DIR_WHERE_IMAGES_ARE = "data/trimmed/*/*"
files=glob(DIR_WHERE_IMAGES_ARE)

def file_size(fn):
    im=Image.open(fn)
    return im.size

# Using a map get all the file sizes
sizes=list(map(file_size,files))

#As thr two lists are same size and same order  - zip
file_size=list(zip(files,sizes))

# I need to extract the Size - which is a tuple as 2 fields
file_size = [[fs[0], fs[1][0], fs[1][1]] for fs in list(zip(files,sizes))]
```

```python
df=pd.DataFrame.from_records(file_size,columns=['filename','X','Y'])
```

```python
df['Class']=df.filename.apply(lambda x: x.split('/')[2])
df['Ratio']=df.X/df.Y
df['Horiz']=df.Ratio.apply(lambda x: True if x < 1.0 else False)
```
